```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

# --- 1. 設定 ---
all_runs = []   # 各回のデータを格納するリスト
labels = []     # 正解ラベル（0:失敗, 1:成功, 2:歌唱）
file_names = [] # 後で確認用にファイル名を保存

print("データ読み込みを開始します...")

# --- 2. 26回分のループ ---
for i in range(1, 27):
    # 前回の分析結果から、ファイルはルート「/」にあると判断
    filename = f"/{i}.txt"

    if os.path.exists(filename):
        try:
            # データの読み込み
            # sep='|' はデータが棒線で区切られているため
            # skipinitialspace=True は余計な空白を消すため
            data = pd.read_csv(filename, sep='|', skipinitialspace=True)

            # 列名の前後にある空白を完全に除去
            data.columns = data.columns.str.strip()

            # 必要な列（ピッチとスティック操作）を抽出
            # ここを増やすことで、速度や高度もAIに教えられます
            features = data[['pitch,__deg', '_elev,stick']].values

            all_runs.append(features)
            file_names.append(f"{i}.txt")

            # --- ラベル付け（ここがあなたの「主観」データ！） ---
            if i == 11:
                labels.append(1) # 11回目：成功（教師データ）
            elif i == 26:
                labels.append(2) # 26回目：歌唱・最後（特殊データ）
            elif i == 1:
                labels.append(0) # 1回目：最初
            else:
                labels.append(0) # その他はとりあえず0に

            print(f"✅ {i}.txt: 読み込み成功 (行数: {len(data)})")

        except Exception as e:
            print(f"❌ {i}.txt: エラー発生 -> {e}")
    else:
        print(f"⚠️ {filename} が見つかりません")

# --- 3. データの形の確認 ---
print("\n" + "="*30)
print(f"読み込み完了！ 合計: {len(all_runs)} 回分")
print("="*30)

# --- 4. 簡単な比較グラフの表示 ---
plt.figure(figsize=(12, 6))
# 1回目（失敗）
plt.plot(all_runs[0][:, 0], label="1st: Panic", color='orange', alpha=0.5)
# 11回目（成功）
plt.plot(all_runs[10][:, 0], label="11th: Success", color='blue', linewidth=2)
# 26回目（歌唱）
plt.plot(all_runs[25][:, 0], label="26th: Singing", color='red', linestyle='--')

plt.title("Comparison: First vs Success vs Singing")
plt.xlabel("Time points")
plt.ylabel("Pitch Angle")
plt.legend()
plt.show()
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# 1. データの長さを揃える (一番長いデータに合わせる)
# all_runs に入っている 26回分のデータを整形
X = pad_sequences(all_runs, dtype='float32', padding='post')
y = np.array(labels)

# 2. データの形を確認 (回数, 時間, 項目数)
print(f"AI用データの形: {X.shape}")

# 3. シンプルなLSTMモデルを構築
model = Sequential([
    # return_sequences=False にすることで、最後の判定結果だけを出す
    LSTM(64, input_shape=(X.shape[1], X.shape[2]), return_sequences=False),
    Dropout(0.2), # 過学習を防ぐおまじない
    Dense(32, activation='relu'),
    Dense(3, activation='softmax') # 3クラス(失敗, 成功, 歌唱)に分類
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
# 学習（Training）の実行
# 26個のデータを50回繰り返し読み込ませます
history = model.fit(
    X, y,
    epochs=50,
    batch_size=4,
    validation_split=0.2, # 20%のデータをテスト用に取っておく
    verbose=1
)

# 学習結果の可視化
plt.figure(figsize=(10, 5))

# 正解率（Accuracy）の推移
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.legend()

# 誤差（Loss）の推移
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.legend()

plt.show()
```

```python
# あなたの「鼻歌（26回目）」をAIに鑑定させる
prediction = model.predict(X[25:26])
result = np.argmax(prediction)
confidence = prediction[0][result] * 100

names = {0: "失敗パターン", 1: "成功パターン", 2: "歌唱（リラックス）パターン"}

print(f"【AI鑑定結果】")
print(f"26回目のフライトは... 「{names[result]}」と判定！")
print(f"AIの自信度: {confidence:.2f}%")
```

```python
# 11回目（成功）と 26回目（歌唱/AIは失敗と判定）のスティック操作を並べてみる
plt.figure(figsize=(12, 5))
plt.plot(all_runs[10][:, 1], label='11th (Success/Strict)', alpha=0.7)
plt.plot(all_runs[25][:, 1], label='26th (Singing/AI said Fail)', color='red', alpha=0.7)
plt.title('Stick Control Comparison: Why AI was so confident?')
plt.legend()
plt.show()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# 1. 11回目（インデックスは10）を基準データとして取り出す
# データを一列に平ら（flatten）にして比較しやすくします
base_data = X[10].reshape(1, -1)

all_scores = []

# 2. 全26回分と11回目を比較する
for i in range(len(X)):
    compare_data = X[i].reshape(1, -1)

    # 類似度（1.0に近いほど似ている）を計算
    score = cosine_similarity(base_data, compare_data)[0][0]
    all_scores.append(score)

# 3. ランキングを整理して表示
print("🏆 【11回目（成功）との類似度ランキング】 🏆")
print("-" * 45)

# スコアが高い順にインデックスを並び替え
ranking_indices = np.argsort(all_scores)[::-1]

for rank, idx in enumerate(ranking_indices):
    # 11回目自身は1位になるので除外して考えたい場合は注目
    medal = "🥇" if rank == 0 else "🥈" if rank == 1 else "🥉" if rank == 2 else "  "

    # 26回目（歌唱回）には目印をつける
    note = " ← 鼻歌回！" if idx == 25 else ""

    print(f"{medal} {rank+1:>2}位: {idx+1:>2}回目 (類似度: {all_scores[idx]:.4f}){note}")

# 4. 最後にちょっとした分析
top_3_idx = ranking_indices[1] # 11回目を除いた実質1位
print("-" * 45)
print(f"💡 11回目の次に『神』だったのは 【 {top_3_idx+1}回目 】 です！")
```
